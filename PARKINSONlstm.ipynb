```
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```
print ('Loading data...')
df = pd.read_csv('/content/drive/MyDrive/major/lstmdatatrial.csv')
#df['sequence'] = df['sequence'].apply(lambda x: [int(e) for e in x.split()])
df = df.reindex(np.random.permutation(df.index))

train_size = int(len(df) * (1 - 0.3))

X_train = df['sequence'].values[:train_size]
y_train = np.array(df['Class Status'].values[:train_size])
X_test = np.array(df['sequence'].values[train_size:])
y_test = np.array(df['Class Status'].values[train_size:])

print(X_train.shape)

   # return pad_sequences(X_train), y_train, pad_sequences(X_test), y_test
```

```
from keras.preprocessing import text, sequence
s = df['sequence'].values[:train_size]
tokenizer = text.Tokenizer(num_words=100,lower=True)
tokenizer.fit_on_texts(s)
seq_token = tokenizer.texts_to_sequences(s)
X_train = sequence.pad_sequences(seq_token, maxlen=16,dtype='float32')

g= df['sequence'].values[train_size:]
tokenizer = text.Tokenizer(num_words=100,lower=True)
tokenizer.fit_on_texts(g)
seq_token = tokenizer.texts_to_sequences(g)
X_test = sequence.pad_sequences(seq_token, maxlen=16,dtype='float32')
```

```
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split
import numpy as np
from keras.layers import Dense, Dropout, LSTM, Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
import pandas as pd
import numpy as np
```

```
print ('Creating model...')
model = Sequential()
model.add(Embedding(input_dim = 188, output_dim = 50, input_length = 192))
model.add(LSTM(256, activation='sigmoid', return_sequences=True))
model.add(Dropout(0.5))
model.add(LSTM(256, activation='sigmoid'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

print ('Compiling...')
model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
```

```
print ('Fitting model...')
hist = model.fit(X_train, y_train, batch_size=64, epochs=200, validation_split = 0.1, verbose = 1)
```

```
score, acc = model.evaluate(X_test, y_test, batch_size=1)
```

```
tf.keras.metrics.FalsePositives(
    thresholds=None, name=None, dtype=None
)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
               metrics=[tf.keras.metrics.FalsePositives()])

model.fit(X_train, y_train, epochs=50, batch_size=1)
test_loss, test_acc = model.evaluate(X_test, y_test)
print('False Positives:', test_acc)
```

```
tf.keras.metrics.FalseNegatives(
    thresholds=None, name=None, dtype=None
)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
               metrics=[tf.keras.metrics.FalseNegatives()])

model.fit(X_train, y_train, epochs=50, batch_size=1)
test_loss, test_acc = model.evaluate(X_test, y_test)
print('FALSE NEGATIVES:', test_acc)
```

```
tf.keras.metrics.TrueNegatives(
    thresholds=None, name=None, dtype=None
)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
               metrics=[tf.keras.metrics.TrueNegatives()])

model.fit(X_train, y_train, epochs=50, batch_size=1)
test_loss, test_acc = model.evaluate(X_test, y_test)
print('TRUE NEGATIVES:', test_acc)
```

```
tf.keras.metrics.TruePositives(
    thresholds=None, name=None, dtype=None
)
model.compile(optimizer='adam',
              loss='binary_crossentropy',
               metrics=[tf.keras.metrics.TruePositives()])

model.fit(X_train, y_train, epochs=50, batch_size=1)
test_loss, test_acc = model.evaluate(X_test, y_test)
print('TRUE POSITIVES:', test_acc)
```
